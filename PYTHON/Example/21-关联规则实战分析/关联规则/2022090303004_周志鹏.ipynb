```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
retail_shopping_basket = {'ID':[1,2,3,4,5,6],
                         'Basket':[['可乐', '炸鸡', '披萨', '薯条', '咖啡'],
                                   ['炸鸡', '可乐', '薯条', '炸鱼', '果汁', '米饭套餐', '牛奶'],
                                   ['苏打水', '薯条', '牛奶'],
                                   ['罗宋汤', '可乐', '炸鸡', '牛奶', '冰淇淋'],
                                   ['苏打水', '红茶', '牛奶', '烤面包'],
                                   ['可乐', '薯条']
                                  ]
                         }
```

```python
retail = pd.DataFrame(retail_shopping_basket)
retail
```

```python
retail_id = retail.drop('Basket' ,1)
retail_id
```

```python
retail_Basket = retail.Basket.str.join(',')
retail_Basket
```

1、数据转换为one-hot编码，并输出

```python
retail_Basket = retail_Basket.str.get_dummies(',')
retail_Basket
```

```python
retail = retail_id.join(retail_Basket)
retail
```

2、提取最小支持度为50%的频繁项集，并输出

```python
frequent_itemsets_2 = apriori(retail.drop('ID',1), min_support=0.50, use_colnames=True)
frequent_itemsets_2
```

3、在上述条件下，提取提升度大于1.1的频繁项集，并输出

```python
rules = association_rules(frequent_itemsets_2, metric='lift', min_threshold=1.1)
rules
```

4、在上述条件下，提取置信度大于0.8的频繁项集，并输出

```python
rules = association_rules(frequent_itemsets_2, metric='confidence', min_threshold=0.8)
rules
```

5.根据上述结果预测最相关性的食品组合

炸鸡和可乐
